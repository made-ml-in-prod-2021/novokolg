```python
data.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholestoral', 'fasting_blood_sugar', 'resting_electrocardiographic_results', 'maximum_heart_rate', 'exercise_induced_angina', 'ST_depression_exercises_vs_rest', 'peak_ST_slope', 'major_vessels_count', 'defect_type', 'target']
```

```python
data.apply(lambda x: len(x.unique()), axis=0)
```

```python
cat_feat =  ['sex', 'chest_pain_type', 'fasting_blood_sugar', 'resting_electrocardiographic_results', 'exercise_induced_angina', 'peak_ST_slope', 'defect_type']
num_feat = ['age', 'resting_blood_pressure', 'cholestoral', 'maximum_heart_rate', 'ST_depression_exercises_vs_rest', 'major_vessels_count']
```

```python
data.corr()
```

```python
g = sns.pairplot(data[num_feat + ['target']], hue="target",
                 hue_order=data.target,
                 palette='Set2')
g.savefig("pairplot.png")
```

```python
data.target.unique()
```

```python
col1 = 'maximum_heart_rate'
col2 = 'ST_depression_exercises_vs_rest'

plt.figure(figsize=(10, 6))

plt.scatter(data[col1][data['target'] == 0],
            data[col2][data['target'] == 0],
            color='red',
            label= 0)

plt.scatter(data[col1][data['target'] == 1],
            data[col2][data['target'] == 1],
            alpha=0.75,
            color='blue',
            label= 1)

plt.xlabel(col1)
plt.ylabel(col2)
plt.legend(loc='best');
```

```python
data.pivot_table('age', 'sex', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("Sex: 1 = male; 0 = female")
plt.ylabel("Target Count")
```

```python
data.pivot_table('sex', 'age', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("Age")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'chest_pain_type', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("chest pain type")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'fasting_blood_sugar', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("Age")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'exercise_induced_angina', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("exercise_induced_angina: 1 = yes, 0 = no")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'peak_ST_slope', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("the slope of the peak exercise ST segment")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'major_vessels_count', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("number of major vessels (0-3) colored by flourosopy")
plt.ylabel("Target Count")
```

```python
data.pivot_table('age', 'defect_type', 'target', 'count').plot(kind='bar', stacked=False)
plt.xlabel("Defect type")
plt.ylabel("Target Count")
```
